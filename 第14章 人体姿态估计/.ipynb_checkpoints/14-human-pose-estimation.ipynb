# 第14章 人体姿态估计

## 14.1 简介

人体姿态估计（human pose estimation）是计算机视觉的一个重要任务，其目的是识别人体的动作和姿势。这对于很多应用来说都非常重要，例如运动分析、虚拟现实和人机交互等。在虚拟现实中，人体姿态识别技术可以帮助跟踪用户的动作，使得虚拟现实体验更加逼真；在人机交互中，人体姿态识别技术可以帮助设备更好地识别用户的命令，从而使得人机交互更加流畅。

一般来说，人体姿态估计任务分为二维姿态估计和三维姿态估计，它们之间有一些明显的差异。
二维姿态估计是指检测人体在图像中的关节点坐标，输出为二维坐标。所以二维姿态估计类似于语义分割和目标检测这种视觉识别任务，所需技术包括图像特征提取和坐标回归等。
三维姿态估计是指识别人体在三维空间中的关节点坐标，输出为三维坐标。所以三维姿态估计通常还需要场景重建技术，如立体视觉、运动估计等。

在本章中，我们主要介绍二维姿态估计，用数学语言描述该任务即是：预定义$N$个关节点，给定$n$张图像，定义图像空间$\mathbb{I}$，给定数据集$\mathcal{D} = \{(\boldsymbol{I}^{(n)},\boldsymbol{y}^{(n)})\}_{n=1}^N$，其中$\boldsymbol{I}^{(n)}\in \mathbb{I}$是数据集中的第$n$张图像，$\boldsymbol{y}^{(n)}=(\boldsymbol{y}^{(n)}_j;j=1 \ldots N)\in\mathbb{R}^{N \times 3}$是这张图像中真实的人体关节点坐标向量，$\boldsymbol{y}^{(n)}_j$是第$j$个关节点的坐标。姿态估计的任务是从$\mathcal{D}$中学习得到一个从图像空间到人体关节点坐标向量空间的映射$f:\mathbb{I} \rightarrow \mathbb{R}^{N \times 3}$，从而给定任意一张测试图像$\boldsymbol{I}$，可以用学习得到的映射函数$f$预测其标签：$\hat{\boldsymbol{y}}=f(\boldsymbol{I})$。由于根据人体结构，可以把关节点连接起来，形成人体骨架（skeleton）的形式（如图14-1所示），所以人体姿势估计也叫做人体骨架估计。
接下来，我们将探讨一些经典的人体姿态估计方法。这些方法已被证明是后续工作的基础，它们的思想在当今仍具有参考价值。

## 14.2 数据集和度量

本节将介绍人体姿态估计相关的数据集和度量。

### 14.2.1 数据集

在人体姿态估计中，通常会选择以下几种常用的数据集进行模型的训练与测试：
1. Leeds Sports Pose (LSP) Dataset $^{[1]}$: 共收录2000张全身姿势图像，这些图像被分为8类（田径、羽毛球、棒球、体操、跑酷、足球、网球和排球），每个人体图像都有14个关节点。
1. Frames Labeled In Cinema (FLIC) Dataset $^{[2]}$: 共收录5003幅图像，这些图像来自30部著名好莱坞电影，每部电影每10帧进行一次人体检测，约20000个人体候选区被发布到众包平台Amazon Mechanical Turk并获得10个上半身关节的真实坐标。最后去除严重遮挡或非正视图的图像。
1. Max Planck Institute for Informatics Human Pose （MPII） Dataset $^{[3]}$: 该数据集包含超过40000张人物图像，涵盖了491种不同的活动。数据集的图像来源于YouTube，涵盖了室内外场景、各种成像条件，因此该数据集在外观变异性方面具有很高的多样性。研究人员对于收集到的图像提供了丰富的注释。注释内容包括身体关节的位置、头部和躯干的3D视角，以及眼睛和鼻子的位置。此外还对所有身体关节和部位的可见性进行了注释。
4. Microsoft Common Objects in Context（MS COCO）Dataset $^{[4]}$：MS COCO数据集不仅可以用于⽬标检测、分割、关键点检测等任务，也可以用于人体姿态估计。MS COCO对每个人体实例都标注了17个关键点，包括鼻子、两眼、两耳、肩部、手肘、手腕、髋部、膝盖和脚踝。相比于MPII数据集，MS COCO在人脸结构的标注上更为丰富和多样化。

### 14.2.2 评测指标

在人体姿态估计中，常常用以下评测指标：
1. Percentage of Correct Parts (PCP)：检测到正确部位的比例。
PCP一般用于衡量肢（limb）是否预测正确，判断标准是：如果预测关节位置与真实关节位置之间的距离小于当前肢体长度的一半，那么就认为这根骨架预测正确。
<!-- 在PCP中认为一个骨架段（limb）被检测到的标准是：骨架段的两个关节点的位置与真实关节点的位置的距离小于骨架段的长度的一半（因此常被写作PCP@0.5）。
    $$
PCP@0.5 = \frac{1}{K} \sum_{i=1}^{K} \frac{\sum_{j=1}^{M} \delta(d_{i,j} < 0.5 L_{i,j})}{M}
    $$ -->
2. Percentage of Correct Keypoints (PCK)：关节点被准确检测的比例。
计算检测的关节点与其对应的真值间的归一化距离小于设定阈值的比例。
3. Object Keypoint Similarity (OKS)：关节点相似度。
具体计算公式为：
    $$
    OKS = \frac{\sum_{i} \exp \left(-d_{i}^{2} / 2 s^{2} k_{i}^{2}\right) \delta\left(v_{i}>0\right)}{\delta\left(v_{i}>0\right)}
    $$
    其中$d_i$是检测到的关节点与相应的真值之间的欧几里得距离，$v_i$是真值的可见性标志，$v_{{i}}=1$表示这个关节点的可见性为1，即关节点无遮挡并且已标注，$v_{i}=2$表示关节点有遮挡但已标注。$s$是尺度因子，其值为人体检测框面积的平方根$\sqrt{H \times W}$，这里的$H, W$是行人检测框的高和宽。
    $k_i$表示该关节点的归一化因子，其是通过对已有数据集中所有的真值计算得到的标准差而获得，反映了当前关节点对整体影响程度。值越大，说明在整个数据集中对该点的标注效果越差；值越小，说明整个数据集中对该点的标注效果越好。
    可以看出，OKS对于关节点的可见性，尺度因子还有关节点标注效果都进行了考虑，因此能够更加综合地评估人体姿态估计的性能。

## 14.3 算法简述

为了更加精确地预测人体的姿态，人们从不同方面进行了大量的探索。
1. 整体视图。例如，关节位置回归、热度图 $^{[6]}$回归。
2. 长距离空间关系。例如，卷积姿态机（convolutional pose machines, CMPs）$^{[7]}$。
3. 迭代估计。例如，迭代误差反馈（iterative error feedback）$^{[8]}$。
4. 网络架构。例如，堆叠沙漏网络（stacked hourglass networks）$^{[9]}$。

接下来我们将介绍一些基于深度学习的人体姿态估计的方法，体会他们在设计时的思考与各自的局限性。

### 14.3.1 DeepPose
正如在简介中介绍的，人体姿态估计的目的是获得映射函数$f$。对于给定的任意一张测试图像$\boldsymbol{I}$，可以用学习得到的映射函数$f$预测其标签：$\hat{\boldsymbol{y}}=f(\boldsymbol{I})$。让我们回忆一下第12章目标检测中，我们详细学习了候选框回归，在这里也用到同样的思想，对关键点的坐标进行回归，就可以估计人体的姿态了。

2014年Alexander Toshev和Christian Szegedy提出的DeepPose $^{[5]}$ 就是秉承着这个想法，这也是首个基于深度学习的人体姿态估计算法。这篇文章构建了一个级联的回归网络，使用深度神经网络来实现人体姿态估计，并将人体姿态估计看作是一个人体关节的回归问题。

如图 14-2 所示，图像的局部信息很难提供有效的姿态信息，作者希望能够采用一种更加全面的方法，构建出一种更通用、更具端到端性质的设计，不需要涉及人体学方面的知识，只要将大量的样例输入，便可获得所需的模型。而正好那几年，计算机视觉领域出现了卷积神经网络这一有潜力的工具，因此便有了DeepPose的出现。

参考文献[5]认为，这种方法有两个优点：（1）深度神经网络具有捕获全局上下文信息的能力，每一个关节的回归器都用到整张图像的信息；（2）比起以往的图模型方法，这种表示方法更加简单，不需要设计人体部分的特征表示或者检测器，也不需要设计人体关节之间的拓扑模型和交互。
<!--   -->
整个DeepPose架构分为初始阶段（initial stage）和后续的精化阶段（refinement stage）。如图 14-3 所示，DeepPose首先将整张图像作为输入，使用卷积神经网络进行全局特征提取，从而初步估计出人体关节点。在这一阶段主要是最小化预测和真实姿态向量之间的$L_2$距离。
$$
\arg \min _{\theta} \sum_{(\boldsymbol{I}, y) \in {D_{N}}}~\sum_{i = 1}^{k}\left\|\boldsymbol{y}_{i}^{(n)}-\psi_{i}(\boldsymbol{I}^{(n)} ; \theta)\right\|_{2}^{2}
$$
上式中，$k$是需要预测的肢体总数，$\theta$是网络模型的参数，$\boldsymbol{y}_{i}^{(n)}$是第$i$个节点的真值姿态向量，$\psi_{i}(\boldsymbol{I}^{(n)} ; \theta)$是网络预测的姿态向量。

由于初始阶段网络的输入尺寸固定为220 $\times$ 220，网络难以兼顾细节信息和全局信息，因此作者设计了精化阶段。在这一阶段，作者通过多阶段优化前一阶段的预测结果，也就是初步的节点坐标，然后在原始图像中根据该坐标选择一定的局部区域作为候选框，达到扩充数据，增加细节的效果，从而提高了回归精度。

总的来说，DeepPose相对于以往的方法有如下创新点：
1. 对于如何使用全局观点获得关节点的坐标的问题，作者提出了使用卷积神经网络作为全局特征提取，将整张图像作为输入的方法。
2. 人体关节点本身存在分布规律，而关节点的坐标在图像坐标系下并无特别的规律，因此作者提出了在每一个环节中，通过中点、长和宽选择一个包围盒，然后计算节点在包围盒中的坐标，从而将绝对坐标转换为统一的坐标。
3. 为了实现更高精度的坐标计算，作者提出了级联的卷积神经网络，在最初输入时，将图像设定为固定大小，通过缩放的方式将图像缩放到这个大小，在获得初步的节点坐标之后，再在原始图像中根据该坐标选择一定的局部区域，从而实现更高准确度的计算节点的坐标。

通过这一系列的设计，DeepPose成功地成为了当时的最高水平（state-of-the-art, SOTA），并引领了深度学习在人体姿态估计领域的潮流。接下来，我们来动手编写DeepPose并在LSP数据集上测试预测效果。

```python
import pickle 
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import random
import scipy.io
from cv2 import copyMakeBorder, flip, BORDER_CONSTANT
from tensorflow.keras.utils import multi_gpu_model 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, MaxPooling2D, 
                                Dense, BatchNormalization, Dropout
```

```python
coords_file_name = "lsp_dataset/joints.mat"
joints = scipy.io.loadmat(coords_file_name)
joints = joints['joints'][:2,:]
W = 220
H = 220
image_list = []
num_image = len(joints[0][0])

for i in range(1, num_image+1):
    if i<10:
        img_name = "im000{0}.jpg".format(i)
    elif i<100:
        img_name = "im00{0}.jpg".format(i)
    elif i<1000:
        img_name = "im0{0}.jpg".format(i)
    else:
        img_name = "im{0}.jpg".format(i)

    img = plt.imread("lsp_dataset/images/{0}".format(img_name))

    if img.shape[0] < H:
        height = H - img.shape[0]
    else:
        height = 0

    if img.shape[1] < W:
        width = W - img.shape[1]
    else:
        width = 0

    if np.mod(width,2) == 0:
        extra_width = 0
    else:
        extra_width = 1

    if np.mod(height,2) == 0:
        extra_height = 0
    else:
        extra_height = 1

    joints[0,:,i-1] = (joints[0,:,i-1] + width/2)/W
    joints[1,:,i-1] = (joints[1,:,i-1] + height/2)/H
    img = copyMakeBorder(img,int(height/2)+extra_height,
                         int(height/2),int(width/2)+extra_width,int(width/2),
                         BORDER_CONSTANT,value=[0,0,0])
    image_list.append(img/255)
```

```python
num_train = 1600
num_validation = num_image-num_train
W = 220
H = 220

X_train = np.array(image_list[0:num_train])
Y_train = np.array(joints[:,:,0:num_train])

X_train = np.reshape(X_train,(num_train,W,H,3))
Y_train = np.transpose(np.reshape(Y_train,(28, num_train)))


X_validation = np.array(image_list[num_train:])
Y_validation = np.array(joints[:,:,num_train:])

X_validation = np.reshape(X_validation,(num_validation,W,H,3))
Y_validation = np.transpose(np.reshape(Y_validation,
                                       (28,num_validation)))
```

### 14.3.2 堆叠沙漏网络

Alejandro Newell等[9]提出使用堆叠沙漏网络来估计人体姿态，该模型引发了业界的关注并受到了广泛的认可，许多后续的人体姿态估计方法都借鉴了这一网络结构。

为了捕捉不同尺度下图像所包含的信息，作者设计了一个复杂的堆叠沙漏模块。局部信息对于特征的提取是必要的，而最终的姿态估计则需要对整体人体的一致理解。从不同尺度下可以提取出许多有用的信息，例如人体的方位、肢体的动作、相邻关节点的关系等。

首先让我们来介绍什么是堆叠沙漏结构：堆叠沙漏结构，顾名思义，就是多个沙漏结构堆叠而成的神经网络。那么什么是沙漏结构呢？
如图14-4 所示，这种堆叠在一起的沙漏模块结构是对称的，每个沙漏结构都分为自下而上（bottom-up）过程和自上而下（top-down）过程。自下而上过程将图像从高分辨率降到低分辨率，自上而下过程将图像从低分辨率升到高分辨率。这种网络结构包含了许多池化和上采样的步骤，池化用于提取特征减少冗余，上采样可以结合多个分辨率的特征。
自下而上先检测图像中人体部件，然后将图像中多人人体的部件分别分配到不同的人体实例上，因此如果只使用自下而上结构，那么在测试推断的时候往往更快速，模型尺寸更小，但因为没能更细致的对人体关节点建模，所以往往准确度稍低。
自上而下过程将人体检测和关节点检测分离，在图像上首先运行一个人体检测器，找到所有的人体实例，对每个人体子图再使用关节点检测。但是只使用自上而下往往比较慢，但姿态估计准确度较高。
将自下而上过程和自上而下过程结合起来，就能将全局信息与局部信息综合起来考虑回归姿态热度图。

网络结构采用了堆叠沙漏模块的形式，从而实现了自底向上和自顶向下的重复过程。作者提出的关键技术是采用中继监督（intermediate supervision）来预测每一个沙漏模块的输出结果，即计算每一个沙漏模块输出的热图损失。需要指出的是，每一个模块的权重是独立的，所有模块都基于相同的真值添加损失函数。

我们先编写沙漏模块的代码：

```python
# 堆叠沙漏模块
class Hourglass(nn.Module):
    def __init__(self, n, channel, increase=0):
        super(Hourglass, self).__init__()
        # 先池化降分辨率
        self.pool = Pool(2, 2)
        
        # 用residual_module来下采样
        self.conv = Residual(channel, channel)
        self.n = n
        
        # 使用递归的方式完成u型结构，每次向内递归一层，
        # 直到n=1不能再套u型层，此时就通过卷积层
        if self.n > 1:
            self.smaller_hourglass = Hourglass(n-1, channel)
        else:
            self.smaller_hourglass = Residual(channel, channel)
            
        # 最近邻上采样，恢复分辨率，对应于self.pool的池化层
        self.up2 = nn.Upsample(scale_factor=2, mode='nearest')

    def forward(self, x):
        
        # 先池化降低分辨率，完成下采样
        pool = self.pool(x)
        # 用residual_module来Downsample
        conv  = self.conv(pool)
        #hourglass,不过参数n=n-1
        low = self.smaller_hourglass(conv)
        # 最近邻上采样
        up  = self.up(low)
        # 将原图像和操作后图像相加，完成u型结构
        return x + up
```

堆叠沙漏网络是基于从模块到子网络再到完整网络的概念设计出来的。一阶沙漏网络是最简单的旁路相加架构，上半路维持原有尺度，下半路经历下采样再进行上采样。而二阶沙漏网络则在一阶架构上嵌入一阶网络。至于之后的高阶沙漏网络，则是一层一层嵌套，从本质上说，子模块都是一样的。
相对于DeepPose直接将人体姿态估计定位成回归问题并将AlexNet改造成对应结构，堆叠沙漏网络显然在姿态估计这个任务上走得更深。堆叠沙漏网络通过多次自下而上过程和自上而下过程来对全局信息和局部信息进行融合，且通过中继监督提升了整个网络的预测精度，同时拥有了能够纠正初始预测错误的能力。如果最初的估计与实际情况相差甚远，那么后续的迭代无法纠正初始的错误。
相较于回归坐标，回归热图的优势有：
* 直接坐标回归，网络需要学习的函数高度非线性，不适合网络学习；
* 如果直接坐标回归，人体特征点坐标的数值范围较大，不适合网络学习；
* 比起坐标的只有两个数，热图能够提供空间上更丰富的监督信息。
